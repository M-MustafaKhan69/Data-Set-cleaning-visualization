```python
s=data["Weight(Kgs)"].astype(str).str.strip().str.lower()
mask_pound=s.str.contains("lbs",na=False)
```

```python
kg_vals=pd.to_numeric(s.str.extract(r"(\d+(?:\.\d+)?)\s*kg",expand=False),errors="coerce")
lbs_vals=pd.to_numeric(s.str.extract(r"(\d+(?:\.\d+)?)\s*lbs",expand=False),errors="coerce")
lbs_to_kg = lbs_vals * 0.45359237
data["Weight(Kgs)"] = kg_vals.fillna(lbs_to_kg)
```

```python
data["Height(Cms)"]=data["Height(Cms)"].astype(str)
height_mask=data["Height(Cms)"].str.contains("'",na=False)
fi=data.loc[height_mask,"Height(Cms)"].str.extract(r"(\d+)'\s*(\d+)")
feet=fi[0].astype(float)
inches=fi[1].astype(float)
data.loc[height_mask,"Height(Cms)"]=(feet*12 + inches)*2.54
```

```python
data["Height(Cms)"]=data["Height(Cms)"].str.replace("cm","",regex=False)
data["Height(Cms)"]=pd.to_numeric(data["Height(Cms)"],errors="coerce")
```
