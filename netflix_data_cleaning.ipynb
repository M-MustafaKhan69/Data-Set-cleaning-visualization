```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
note= pd.read_csv("C:\\Users\\129mu\\Desktop\\data_cleaning_yt\\netflix_titles_raw.csv",encoding="latin1")
note
```

```python
note[["director","cast","country"]]=note[["director","cast","country"]].fillna("N/A")
```

```python
note
```

```python
note["show_id"]=note["show_id"].astype("str").str.replace(r"[A-za-z]","S",regex="True")
```

```python
note["rating"].unique()
```

```python
note["rating"]=note["rating"].replace({
"G": "General Audience",
    "PG": "Parental Guidance Suggested",
    "PG-13": "Parents Strongly Cautioned (Under 13)",
    "R": "Restricted",
    "NC-17": "No Children Under 17",
    "NR": "Not Rated",
    "UR": "Unrated",

    "TV-Y": "Young Children",
    "TV-Y7": "Children 7+",
    "TV-Y7-FV": "Children 7+ (Fantasy Violence)",
    "TV-G": "General Audience",
    "TV-PG": "Parental Guidance Suggested",
    "TV-14": "Parents Strongly Cautioned (14+)",
    "TV-MA": "Mature Audience Only"
})
```

```python
note.loc[note["rating"].str.contains("min", na=False), "rating_full"] = None
```

```python
note.loc[note["rating"].str.contains("min",na=False),"rating"]=None
```

```python
note["rating"].unique()
```

```python
note["rating"]=note["rating"].fillna("Not Rated")
```

```python
note=note.drop(columns=["rating_full"])
```

```python
types= note["Type"].value_counts()
print(types)
```

```python
plt.Figure(figsize=(6,4))
plt.bar(types.index,types.values,color=['red','blue'],label="Types of shows Distribution")
plt.legend()
plt.title("Number of Tv shows on Netflix")
plt.xlabel("Type of Show")
plt.ylabel("Count")
plt.tight_layout()


```

```python
print("Rating count visualisation")

rating=note["Rating"].value_counts()

plt.Figure(figsize=(8,6))
plt.pie(rating,labels=rating.index,autopct="%1.1f%%",pctdistance=0.9,labeldistance=1.2)
plt.title("Rating Distribution chart for Netflix Tv shows")
plt.tight_layout()
plt.show()






```

```python
print("Mins visualization through histogram:")

movies = note[note['Type'] == 'Movie'].copy()

movies = movies.dropna(subset=['Duration(Mins)'])

plt.figure(figsize=(8,6))
plt.hist(movies['Duration(Mins)'], bins=30, color='red', edgecolor='black')

plt.title("Movie Duration Distribution")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")

plt.tight_layout()
plt.show()



```

```python
print("Relation b/w released year and shows/Movies")

release_count =note['Release_year'].value_counts().sort_index()


plt.Figure(figsize=(10,8))
plt.scatter(release_count.index,release_count.values,color='gold',marker='d')
plt.xlabel("Released year")
plt.ylabel("Count of Movies/Shows")
plt.title("No. of Movies/Shows released each year")
plt.tight_layout()
plt.show()





```

```python
print("Top 10 countries with most Shows & Movies")

country_count= note["Country"].value_counts().head(10)

plt.Figure(figsize=(10,8))

plt.barh(country_count.index, country_count.values, color="#34495E",label="Top 10 countries")
plt.title("Countries with Most Movie/Tv-Show")
plt.ylabel("Country")
plt.xlabel("Number of Shows/Movies")
plt.legend()
plt.tight_layout()
plt.show()







```

```python
fig,ax=plt.subplots(1,2,figsize=(8,6))




content_year= note.groupby(['Release_year',"Type"]).size().unstack()

ax[0].scatter(content_year.index,content_year["Movie"],color='green', marker='d')
ax[0].set_title("Movies Release Per year")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Number Of Movies")

ax[1].scatter(content_year.index,content_year["TV Show"],color='orange',marker='^')
ax[1].set_title("TV Shows Release Per year")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("Number Of TV Shows")

fig.suptitle("Comparison Of Movies & TV Shows Released Per Pear")

plt.tight_layout()
plt.show()











```
