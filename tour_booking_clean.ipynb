```python
import unicodedata

inf["Artist"]=inf["Artist"].apply(lambda x: unicodedata.normalize("NFKD",str(x)))
inf["Artist"]=(inf["Artist"].str.encode("ascii","ignore").str.decode("utf-8"))
inf["Artist"]=inf["Artist"].str.replace(r"[^\w\s]+","",regex=True)
inf

```
